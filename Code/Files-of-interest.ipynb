# This notebook is used to extract all erroneous files from Dominique's Benchmark

```python
from pathlib import Path
from collections import defaultdict
import json
import os


base_dir = Path("//nas-01/TrainingsDataExt4/steady-faster-learning/confidence_score")

versions = ["gpt-4.1","gpt-4.1-mini","gpt-5","gpt-5-mini","gpt-5-nano"]

for version in versions:
    special_dir = base_dir / version
    files = sorted(special_dir.rglob("*.pdf"))

    occurence_dict = defaultdict(int)

    for file in files:
        occurence_dict[file.stem] += 1

    sorted_occurrences = dict(sorted(occurence_dict.items(), key=lambda x: x[1], reverse=True))

    print(sorted_occurrences)

    output_dir = Path(f"../Files-of-interest")
    output_dir.mkdir(exist_ok=True)
    with open(output_dir / f"{version}-output.json", "w", encoding="utf-8") as f:
        json.dump(occurence_dict, f, indent=2, ensure_ascii=False)


files = sorted(base_dir.rglob("*.pdf"))

occurence_dict = defaultdict(int)

for file in files:
    occurence_dict[file.stem] += 1

sorted_occurrences = dict(sorted(occurence_dict.items(), key=lambda x: x[1], reverse=True))

print(sorted_occurrences)

output_dir = Path(f"../Files-of-interest")
output_dir.mkdir(exist_ok=True)
with open(output_dir / f"combined-output.json", "w", encoding="utf-8") as f:
    json.dump(occurence_dict, f, indent=2, ensure_ascii=False)

```
